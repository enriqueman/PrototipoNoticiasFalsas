```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
categorias = ['Estoy satisfecho con el sistema', 'Simple de usar', 'Cómodo de usar', 'Fácil de aprender a usar', 'Rápido de entender su finalidad', 'Capaz de proporcionar las ayudas', 'Altamente placentero en su uso', 'Agradable de usar']
valores = [92.48, 66.7, 64.9, 63.2, 57.9, 52.6, 63.2, 63.2]

num_vars = len(categorias)

# Calcula los ángulos
angulos = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Los gráficos de radar se repiten en el primer valor para cerrar el círculo
valores += valores[:1]
angulos += angulos[:1]
categorias += categorias[:1]  # Agrega la primera etiqueta al final

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

ax.fill(angulos, valores, color='red', alpha=0.25)
ax.set_yticklabels([])
ax.set_xticks(angulos)
ax.set_xticklabels(categorias)

# Cambia la escala del eje radial para que vaya de 0 a 100
ax.set_ylim(0, 100)

# Agrega anotaciones con los porcentajes
for i in range(num_vars):
    angle_rad = i / float(num_vars) * 2 * np.pi
    if angle_rad == 0:
        ha, distance_ax = "center", 10
    elif 0 < angle_rad < np.pi:
        ha, distance_ax = "left", 1
    elif angle_rad == np.pi:
        ha, distance_ax = "center", 1
    else:
        ha, distance_ax = "right", 1
    ax.text(angle_rad, 100 + distance_ax, str(valores[i])+'%', size=10, horizontalalignment=ha, verticalalignment="center")

plt.show()
```
